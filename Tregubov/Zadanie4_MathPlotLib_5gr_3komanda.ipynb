```python
# копия mathlab
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure()

plt.subplot(211) # или так: plt.subplot(2, 1, 1)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
#Логарифмические и нелинейные оси координат
np.random.seed(19680801)
# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))
# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')  # маштаб 
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)
# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# настройка графиков https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yscale.html#matplotlib.pyplot.yscale
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.65,wspace=0.75)
plt.show()
```

```python
# подобие словаря, только индексы могут быть одинаковыми
ts = pd.Series(np.random.randn(1000), index = pd.date_range("1/1/2000", periods=1000)) #https://pyprog.pro/pd/pd_10_data_structures.html
ts = ts.cumsum() #возвращает кумулятивную (накапливаемую) сумму элементов массива
ts.plot()
plt.show()
```

```python
#Самая простая аналогия для DataFrame - это таблица, столбцами которой являются объекты Series.
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=ts.index,  # указываем нужные индексы:
                  columns=list("ABCD"))  # указываем необходимые столбцы:
df = df.cumsum()
plt.figure()
df.plot()
plt.show()
```

```python
# таким образом выводим
imgplot = plt.imshow(img)
```

```python
# mpimg может быть полезным инструментом для повышения контрастности и более легкой визуализации данных
# мы можем просто выбрать один канал наших данных:
lum_img = img[:, :, 0]
plt.imshow(lum_img);
```

```python
# Чаще всего "интересная" часть изображения находится вокруг пика, и вы можете получить дополнительный контраст,
# обрезав области выше и/или ниже пика. В нашей гистограмме, похоже, не так много полезной информаци, но все же
imgplot = plt.imshow(lum_img, clim=(0.0, 0.2))
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(lum_img)
ax.set_title('Before')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(lum_img)
imgplot.set_clim(0.0, 0.7)
ax.set_title('After')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
```

```python
#Схемы интерполяции массивов
from PIL import Image

#Интерполяция-это то, как вы заполняете это пространство.
# когда мы увиличиваем изображение, Старых пикселей больше нет, и компьютер должен рисовать пиксели,
# чтобы заполнить это пространство.

#Здесь у нас есть интерполяция по умолчанию, билинейная, так как мы не дали imshow() любой аргумент интерполяции.
img = Image.open('Mikhail-Shcheglov.png')
img.thumbnail((64,64 ), Image.ANTIALIAS)  # изменяет размер изображения на месте
imgplot = plt.imshow(img)
```

```python
# Давайте попробуем некоторые другие. Вот "ближайший", который не делает интерполяции.
imgplot = plt.imshow(img, interpolation="nearest")
```

```python
# и бикубический:
imgplot = plt.imshow(img, interpolation="bicubic")
```
